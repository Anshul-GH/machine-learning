# Lambda Functions


```python
def addition(a,b):
    return a+b
```

```python
addition(4,5)
```

```python
addition = lambda a,b:a+b
```

```python
addition(12,14)
```

```python
def even(num):
    if num % 2 == 0:
        return True
```

```python
even(24)
```

```python
even1 = lambda x: x % 2 == 0
```

```python
even1(12)
```

```python
even1(13)
```

```python
def add_three(x,y,z):
    return x + y +z
```

```python
add_lambda = lambda x,y,z: x+y+z
```

```python
add_lambda(1,2,3)
```

```python
add_three(1,2,3)
```

```python
add_args = lambda *x: sum(x)
```

```python
add_args(1,2,3,4,5)
```

# Map Function

```python
def even_or_odd(num):
    if num % 2 == 0:
        return f"The number {num} is Even"
    else:
        return f"The number {num} is Odd"
```

```python
even_or_odd(12)
```

```python
even_or_odd(13)
```

```python
# applying for multiple numbers

lst = [1,2,3,4,5,6,7,8,9]
```

```python
out = list(map(even_or_odd, lst))
out
```

```python
out
```

# Filter Functions

```python
def even(num):
    if num % 2 == 0:
        return True
```

```python
lst = [1,2,3,4,5,6,7,8,9,0]
list(filter(even, lst))
```

```python
list(filter(lambda num: num % 2 == 0, lst))
```

```python
list(map(lambda num: num % 2 == 0, lst))
```

## Univariate Analysis

```python
df_setosa = df.loc[df['species'] == 'Iris-setosa']
df_versicolor = df.loc[df['species'] == 'Iris-versicolor']
df_virginica = df.loc[df['species'] == 'Iris-virginica']
```
